```python
import pandas as pd

# Load CSV
CSV_FILE = "results.csv"
df = pd.read_csv(CSV_FILE)

# Group by columns, but take max diff instead of mean
grouped = (
    df.groupby(["graph_type", "model", "cascade_size_limit", "p_infect", "centrality"])
      .agg(min_diff=("diff", "min"), min_idx=("diff", "idxmin"))
      .reset_index()
)

# Now filter for IC model and k-regular graph
df_filtered = grouped[
    (grouped["model"] == "IC") &
    (grouped["graph_type"] == "k_regular") &
    (grouped["cascade_size_limit"] == 100)
]

if not df_filtered.empty:
    # Row with maximum diff
    max_row = df_filtered.loc[df_filtered["min_diff"].idxmin()]

    # Look up the id from the original df
    max_id = df.loc[max_row["min_idx"], "id"]

    print("Max diff:", max_row["min_diff"])
    print("ID of row with max diff:", max_id)
    display(max_row)
else:
    print("No data found for IC / k-regular / cascade 100")
```
