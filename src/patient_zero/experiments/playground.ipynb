```python
import pandas as pd

df = pd.read_csv("results.csv")

df = df[df["centrality"].isin(["rumor_centrality", "distance_centrality"])]

pivot_df = df.pivot_table(
    index=[
        "graph_type",
        "model",
    ],
    columns="centrality",
    values="estimate_error",
    aggfunc="mean"
).reset_index()

pivot_df["error_diff"] = pivot_df["rumor_centrality"] - pivot_df["distance_centrality"]

def get_winner(x):
    if x < 0:
        return "Rumor"
    elif x > 0:
        return "Distance"
    return "Tie"

overall = pivot_df.groupby(["graph_type", "model"])["error_diff"].mean().reset_index()
overall["winner"] = overall["error_diff"].apply(get_winner)

overall
```
